```python
# fill in the information here
n_rows = 360 
n_cols = 240
nR=88
```

```python
t_file =os.path.join(ts_folder, 'ts_hour.0000027000.data')
t_grid = np.fromfile(t_file, '>f4').reshape((2,nR,n_rows,n_cols))

vel_file =os.path.join(vel_folder, 'vel_hour.0000027000.data')
vel_grid = np.fromfile(vel_file, '>f4').reshape((3,nR,n_rows,n_cols))
```

```python
fig = plt.figure(figsize = (10,6))
C = plt.pcolormesh(t_grid[0,0,:,:], cmap=cm.thermal,vmin=10,vmax=20)
plt.colorbar(C)
plt.xlabel('Columns')
plt.ylabel('Rows')
#plt.gca().invert_yaxis()
plt.title(' cross-section')
plt.axhline(y=200, color='red', linestyle='--', linewidth=1.5)
plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
C = plt.pcolormesh(vel_grid[0,0,:,:], cmap='RdBu',vmin=-1,vmax=1)
plt.colorbar(C)
plt.xlabel('Columns')
plt.ylabel('Rows')
```

```python
fig = plt.figure(figsize = (10,6))
C = plt.pcolormesh(t_grid[0,:,180,:], cmap=cm.thermal,vmin=0,vmax=17)
plt.colorbar(C)
plt.xlabel('Columns')
plt.ylabel('Vertical level')
plt.gca().invert_yaxis()
plt.title(' cross-section')
plt.show()
```
