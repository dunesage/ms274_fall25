```python
# fill in the information here
n_rows = 360 
n_cols = 240
nR=88
```

```python
# read in the file and reshape it to the proper dimensions
t_file = os.path.join(lesson_folder, 'T.0000000000.data')

t_grid = np.fromfile(t_file, '>f4').reshape((nR,n_rows,n_cols))

s_file = os.path.join(lesson_folder, 'S.0000000000.data')

s_grid = np.fromfile(s_file, '>f4').reshape((nR,n_rows,n_cols))

u_file = os.path.join(lesson_folder, 'U.0000000000.data')

u_grid = np.fromfile(u_file, '>f4').reshape((nR,n_rows,n_cols))

v_file = os.path.join(lesson_folder, 'V.0000000000.data')

v_grid = np.fromfile(v_file, '>f4').reshape((nR,n_rows,n_cols))

# read in the file and reshape it to the proper dimensions

hfacc = os.path.join(lesson_folder, 'hFacC.data')

h_grid = np.fromfile(hfacc, '>f4').reshape((90,n_rows,n_cols))
```

Next, let's plot the data on a figure with a colorbar:

```python
for i in range(n_cols):
    for j in range(n_rows):
        for k in range(nR):
            if h_grid[k,j,i]>0:
                if t_grid[k,j,i]==0:
                    t_grid[k+1,j,i]=t_grid[k,j,i]

for i in range(n_cols):
    for j in range(n_rows):
        for k in range(nR):
            if h_grid[k,j,i]>0:
                if s_grid[k,j,i]==0:
                    s_grid[k+1,j,i]=s_grid[k,j,i]

for i in range(n_cols):
    for j in range(n_rows):
        for k in range(nR):
            if h_grid[k,j,i]>0:
                if u_grid[k,j,i]==0:
                    u_grid[k+1,j,i]=u_grid[k,j,i]

for i in range(n_cols):
    for j in range(n_rows):
        for k in range(nR):
            if h_grid[k,j,i]>0:
                if v_grid[k,j,i]==0:
                    v_grid[k+1,j,i]=v_grid[k,j,i]
```

```python
init_grids = [t_grid, s_grid, u_grid, v_grid]

variable_names=['THETA','SALT','UVEL','VVEL']

for variable_name, init_grid in zip(variable_names, init_grids):
    output_file = os.path.join(lesson_folder, 'newICs/', variable_name + '_IC.bin')
    init_grid.astype('>f4').tofile(output_file)
```

```python
boundaries = ['north', 'south', 'west']
for variable_name, init_grid in zip(variable_names, init_grids):
    for boundary in boundaries:
        if boundary == 'north':
                bc_data = init_grid[:, -1,:]
        elif boundary == 'south':
                bc_data = init_grid[:, 0,:]
        elif boundary == 'west':
                bc_data = init_grid[:, :,0]
        print(np.shape(bc_data))
        bc_data_ext = np.zeros((12, bc_data.shape[0], bc_data.shape[1]))
        print(np.shape(bc_data_ext))
        for i in range(12):
            bc_data_ext[i,:,:] = bc_data

        bc_data=bc_data_ext
        output_file = os.path.join(lesson_folder, 'newBCs/',f'{variable_name}_{boundary}_2011')
        
        # Write file
        bc_data.astype('>f4').tofile(output_file)
```
