### Part 1: Horizonal Velocity Structure
First, examine the mean velocity on the west and east coasts of the U.S. Begin by loading in both the `EVEL` and `NVEL` fields from the `EVEL_2015_12.nc` and `NVEL_2015_12.nc` files, respectively -- these fields represent the zonal (eastward-flowing) and meridional (northward-flowing) velocity in the model. 

```python
# load in datasets
ds_evel=xr.open_dataset('EVEL_2015_12.nc')
ds_nvel=xr.open_dataset('NVEL_2015_12.nc')

# set array variables
evel=np.array(ds_evel['EVEL'])
nvel=np.array(ds_nvel['NVEL'])

# save coords
longitude=np.array(ds_evel['longitude'])
latitude=np.array(ds_evel['latitude'])
```

Now that you have the data read in, make a plot of the eastward and northward velocity at the surface of the ocean. Be sure to add a title to each plot and some labels for the x and y axes.

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.title('Eastward velocity')
plt.pcolormesh(longitude,latitude,evel[0,0,:,:],vmin=-1,vmax=1,cmap='seismic')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Eastward velocity (m/s)',orientation='horizontal')

plt.subplot(1,2,2)
plt.title('Northward velocity')
plt.pcolormesh(longitude,latitude,nvel[0,0,:,:],vmin=-1,vmax=1,cmap='seismic')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Northward velocity (m/s)',orientation='horizontal')
plt.show()
```

### Part 3: Vertical Structure

```python
# load in dataset, set array variables
ds_wvelstar=xr.open_dataset('WVELSTAR_2015_12.nc')
wvelstar=np.array(ds_wvelstar['WVELSTAR'])
depth=np.array(ds_wvelstar['Zl'])
```

```python
fig = plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
C1 = plt.pcolormesh(longitude,depth,theta[0,:,259,:], vmin = -2, vmax=30, cmap='turbo')
plt.colorbar(C1, label='Temperature (C)' )
plt.title('Temperature')
plt.ylabel('Depth (m)')
plt.xlabel('Longitude');

plt.subplot(1,2,2)
C2 = plt.pcolormesh(longitude,depth,wvelstar[0,:,259,:],vmin = -10**-5, vmax=10**-5, cmap='RdBu')
plt.colorbar(C2, label='Vertical velocity (m/s)' )
plt.title('Vertical velocity')
plt.ylabel('Depth (m)')
plt.xlabel('Longitude');
```
